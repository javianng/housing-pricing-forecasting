## Adding the SORA values into the HDB working dataset

```python
import pandas as pd
```

```python
sora_data = pd.read_csv('../data/modified/3M_SORA_rates_2005_2024_cleaned.csv')
sora_data.head()
```

We will convert the data into a dictionary so that we can easily add it into the HDB dataframe later on.

```python
sora_dict = {}
for (index, row) in sora_data.iterrows():
    sora_dict[row.iloc[0]] = row.iloc[1]
```

```python
hdb_data = pd.read_csv('../data/modified/hdb_working_data.csv')
hdb_data.head()
```

Now we add the SORA values to the dataframe based on the transaction month.

```python
col_list = hdb_data.columns.tolist()
col_list.append('sora')
output_data = pd.DataFrame(columns=col_list)
for (index, row) in hdb_data.iterrows():
    output_row = row.tolist()
    output_row.append(sora_dict.get(row.iloc[0]))
    output_data.loc[len(output_data.index)] = output_row
output_data.head()
```

```python
output_data.to_csv('../data/modified/hdb_working_data.csv', index=False)
```
